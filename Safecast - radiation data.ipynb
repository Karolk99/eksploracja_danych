## 1. Czyszczenie danych

```python
from large_data_fitler import LargeDataFilter
from chunk_filters import (
    DropColumnsFilter,
    SpecificValuesInRowFilter,
    EmptyCellsFilter,
    ConvertToDateTimeFilter,
)

input_file = "measurements-out.csv"
output_file = "cleaned_data.csv"
tmp_file = "temporary.csv"
tmp_file2 = "temporary2.csv"
tmp_file3 = "temporary3.csv"
tmp_file4 = "temporary4.csv"
```

Usuwamy wszystkie dane, które mają inne jednostki niż "cpm".

```python
filter = SpecificValuesInRowFilter("Unit", ["cpm"])
LargeDataFilter().filter_data(
    input_file,
    tmp_file,
    filter,
)
```

Usuwamy wszystkie kolumny, które są dla nas zbędne. Bardzo istotne jest jak najdokładniejsze zawężenie danych ze względu na ich duży rozmiar.

```python
columns_to_drop = [
    "Unit",
    "Location Name",
    "MD5Sum",
    "Surface",
    "Radiation",
    "Uploaded Time",
    "Loader ID",
]
filter = DropColumnsFilter(columns_to_drop)
LargeDataFilter().filter_data(
    tmp_file,
    tmp_file2,
    filter,
)
os.remove(tmp_file)
```

Usuwamy wszystkie kolumny, które nie mają wartości w czterech najważniejszych kolumnach

```python
columns = ["Captured Time", "Latitude", "Longitude", "Value"]
filter = EmptyCellsFilter(columns)
LargeDataFilter().filter_data(
    tmp_file2,
    tmp_file3,
    filter,
)
os.remove(tmp_file2)
```

Zamieniamy kolumnę "Captured Time" na standardowy format w celu ułatwienia dalszej obróbki.

```python
filter = ConvertToDateTimeFilter("Captured Time", errors="coerce")
LargeDataFilter().filter_data(
    tmp_file3,
    tmp_file4,
    filter
)
os.remove(tmp_file3)
```

Usuwamy te wiersze w których, nie udało się ustandaryzować formatu timestampa.

```python
columns = ["Captured Time"]
filter = EmptyCellsFilter(columns)
LargeDataFilter().filter_data(
    tmp_file4,
    output_file,
    filter,
)
os.remove(tmp_file4)
```

## 2. Wizualizacja przykładowych urządzeń

```python
from chunk_filters import DistinctValuesFilter

input_file = "cleaned_data.csv"
output_file = "list_of_devices.csv"

filter = DistinctValuesFilter(["Device ID"])
LargeDataFilter().filter_distinct(
    input_file,
    output_file,
    filter,
)
```

```python
import plotly.express as px

df = pd.read_csv(output_file)
df["Height"] = df["Height"].fillna("unknown")

fig = px.scatter_mapbox(
    df,
    lat="Latitude",
    lon="Longitude",
    hover_name="Device ID",
    hover_data=["Height"],
    color_discrete_sequence=["fuchsia"],
    zoom=3,
    height=900,
)
fig.update_layout(
    mapbox_style="white-bg",
    mapbox_layers=[
        {
            "below": "traces",
            "sourcetype": "raster",
            "sourceattribution": "United States Geological Survey",
            "source": [
                "https://basemap.nationalmap.gov/arcgis/rest/services/USGSImageryOnly/MapServer/tile/{z}/{y}/{x}"
            ],
        }
    ],
)
fig.update_layout(mapbox_style="open-street-map")
# fig.update_layout(mapbox_style="dark", mapbox_accesstoken=token)
# fig.update_layout(margin={"r": 0, "t": 0, "l": 0, "b": 0})
fig.show()
```
