```python
import sys

sys.path.insert(0, 'filters')

import pandas as pd

from large_data_fitler import LargeDataFilter
from chunk_filters import (
    SpecificValuesInRowFilter,
    EmptyCellsFilter,
    ConvertToDateFilter,
    DropColumnsFilter,
    InRadiusFilter,
)
```

# Read and filter data

```python
input_file = 'measurements.csv'
output_file = 'filtered_columns.csv'

columns_to_drop = [
    "Location Name",
    "MD5Sum",
    "Surface",
    "Radiation",
    "Uploaded Time",
    "Loader ID",
]
chunk_filter = DropColumnsFilter(columns_to_drop)

LargeDataFilter().filter_data(
    input_file,
    output_file,
    chunk_filter,
)
```

```python
pd.read_csv(output_file, nrows=10)
```

```python
input_file = 'filtered_columns.csv'
output_file = 'filtered_unit.csv'

cpm_filter = SpecificValuesInRowFilter(column='Unit', acceptable_values=['cpm'])

LargeDataFilter().filter_data(
    input_file,
    output_file,
    cpm_filter,
)

pd.read_csv(output_file, nrows=10)
```

```python
input_file = 'filtered_unit.csv'
output_file = 'in_radius.csv'

power_plant_coordinates = (37.421389, 141.0325)
in_radius_filter = InRadiusFilter(radius=200, power_plant_coordinates=power_plant_coordinates)

LargeDataFilter().filter_data(
    input_file,
    output_file,
    in_radius_filter,
)

pd.read_csv(output_file, nrows=10)
```

```python
input_file = 'in_radius.csv'
output_file = 'in_radius_drop_nan.csv'

empty_cell_filter = EmptyCellsFilter(['Captured Time', 'Value'])

LargeDataFilter().filter_data(
    input_file,
    output_file,
    empty_cell_filter,
)

pd.read_csv(output_file, nrows=10)
```

```python
input_file = 'in_radius_drop_nan.csv'
output_file = 'removed_hours.csv'

convert_to_date_filter = ConvertToDateFilter('Captured Time')

LargeDataFilter().filter_data(
    input_file,
    output_file,
    convert_to_date_filter,
)

pd.read_csv(output_file, nrows=10)
```

# Radiation Time Series in 200 km radius from power plant

```python
import dask.dataframe as dd
from dask.diagnostics import ProgressBar

ProgressBar().register()

input_file = 'removed_hours.csv'

ddf = dd.read_csv(input_file)

df = ddf.groupby('Captured Time').mean().compute().reset_index()
df = df[pd.to_datetime(df['Captured Time']) < pd.to_datetime("today")]

df
```
